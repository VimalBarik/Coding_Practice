**Be careful when copying arrays**

```python
a = np.array([1,2,23])
b = a #variable b points to the same thing as a does 
b[0] = 100
print(a)
```

```python
# to prevent this we use .copy() function
c = a.copy()
c[0] = 200
print(a)
print(c)
```

#### Reorganizing arrays

```python
before = np.array([[1,2,3,4], [5,6,7,8]])
print(before)

after = before.reshape((2,2,2))
print(after)
```

```python
# Vertically stacking vectors
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.vstack([v1,v2])
np.vstack([v1,v2,v1,v2])
```

```python
# Horizontal stack
h1 = np.ones((2,4))
h2 = np.zeros((2,2))

np.hstack((h1, h2))
```
