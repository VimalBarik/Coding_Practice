```python
def getAlternates(arr):
  res = []

  #Iterate over all alternate elements
  for i in range (0, len(arr), 2):
    res.append(arr[i])

  return res

arr = [10, 20, 30, 40, 50]

print(getAlternates(arr))
```

```python
# Function to find the leaders in an array
def leaders(arr):
  result = []
  n = len(arr)

  for i in range(n):
    for j in range(i + 1, n):
      if arr[i] < arr[j]:
        break
    else:
      result.append(arr[i])

  return result


arr = [16, 17, 4, 3, 5, 2]

print(leaders(arr))
```

```python
def remove_duplicates(arr):

  seen = set()

  idx = 0

  for i in range(len(arr)):
    if arr[i] not in seen:
      seen.add(arr[i])

  return seen

arr = [1, 2, 2, 3, 4, 4, 4, 5, 5]

print(remove_duplicates(arr))
```

```python
#Prints all subarrays in arr[0..n-1]

def sub_array(arr):
  n = len(arr)

  for i in range(n):
    for j in range(i, n):
      for k in range(i, j + 1):
        print(arr[k], end='')
      print()

arr = [1, 2, 3, 4]
print(sub_array(arr))
```

```python
def reverseArray(arr):
  n = len(arr)

  temp = [0]*n

  for i in range(n):
    temp[i] = arr[n-i-1]

  for i in range(n):
    arr[i] = temp[i]

  return arr


arr = [1, 4, 3, 2, 6, 5]

print(reverseArray(arr))
```

```python
def reverseArray(arr):
  left = 0
  right = len(arr) - 1


  while left < right:
    arr[left], arr[right] = arr[right], arr[left]

    left += 1
    right -= 1
  return arr

arr = [1, 4, 3, 7, 2, 6, 5]

print(reverseArray(arr))
```

```python
def reverseArray(arr):

  n = len(arr)

  for i in range(n//2):
    temp = arr[i]
    arr[i] = arr[n-i-1]
    arr[n-i-1] = temp

  return arr


arr = [1, 4, 3, 7, 2, 6, 5]
print(reverseArray(arr))
```
