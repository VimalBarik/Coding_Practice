```python
import tensorflow as tf
```

```python
mnist = tf.keras.datasets.mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_train, X_test = X_train/255.0, X_test/255.0
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Input, Dropout

model = Sequential()
model.add(Input((28, 28)))
model.add(Flatten())
model.add(Dense(128, activation="relu"))
model.add(Dense(64, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(10))
```

```python
nonsense_prediction = model(X_train[:1]).numpy()
nonsense_prediction
```

```python
tf.nn.softmax(nonsense_prediction).numpy()
```

```python
from tensorflow.keras.losses import SparseCategoricalCrossentropy

loss_fn = SparseCategoricalCrossentropy(from_logits = True)

loss_fn(y_train[:1], nonsense_prediction)
```

```python
from tensorflow.keras.optimizers import Adam

optimizer = Adam(learning_rate= 0.01)

model.compile(optimizer=optimizer, loss = loss_fn, metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10)
```

```python
model.evaluate(X_test, y_test)
```

```python
import tensorflow as tf
import numpy as np

np.random.seed(42)
X = np.linspace(0, 1, 100).astype(np.float32)
y = 2 * X + 1 + np.random.normal(0, 0.1, 100).astype(np.float32)

w = tf.Variable(0.0)
b = tf.Variable(0.0)

def predict(x):
  return w * x + b

def mse_loss(y_pred, y_true):
  return tf.reduce_mean((y_pred-y_true)**2)

learning_rate = 0.1
for epoch in range(200):
  with tf.GradientTape() as tape:
    y_pred = predict(X)
    loss = mse_loss(y_pred, y)

  grads = tape.gradient(loss, [w, b])
  w.assign_sub(learning_rate * grads[0])
  b.assign_sub(learning_rate * grads[1])

  if epoch % 40 == 0:
    print(f"Epoch {epoch}: loss={loss.numpy():.4f}, w={w.numpy():.3f}, b={b.numpy():.3f}")



print(f"\nLearned: w={w.numpy():.3f}, b={b.numpy():.3f}")
```

```python
class MyMLP(tf.keras.Model):
  def __init__(self):
    self.dense1 = tf.keras.layers.Dense(64, activation='relu')
    self.dense2 = tf.keras.layers.Dense(32, activation='relu')
    self.out = tf.keras.layers.Dense(10, activation='softmax')
    self.dropout = tf.keras.layers.Dropout(0.3)

  def call(self, inputs, training=False):
    x = self.dense1(inputs)
    x = self.dropout(x, training=training)
    x = self.dense2(x)
    return self.out(x)

model = MyMLP()
dummy = tf.random.normal([4, 20])
out = model(dummy, training=True)
print(out.shape)
```
