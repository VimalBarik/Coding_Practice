```python
import pandas as pd
```

```python
airplane = pd.read_csv('aviation_accidents_detailed.csv')
```

```python
airplane.head()
```

```python
airplane[['Fatalities', 'Occupants']] = airplane['Fatalities'].str.extract(r'Fatalities:\s*(\d+)\s*/\s*Occupants:\s*(\d+)', expand=True)

# Convert them to numeric types
airplane['Fatalities'] = pd.to_numeric(airplane['Fatalities'], errors='coerce')
airplane['Occupants'] = pd.to_numeric(airplane['Occupants'], errors='coerce')
```

```python
airplane
```

```python
airplane.info()
```

```python
airplane['Date'] = pd.to_datetime(airplane['Date'], errors='coerce')
```

```python
airplane = airplane.sort_values(by='Date')
```

```python
airplane.tail()
```

```python
airplane.info()
```

```python
airplane = airplane.dropna(subset=['Date'])
```
